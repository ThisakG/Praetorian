```python
import joblib
from pprint import pprint

# Load model
model = joblib.load("../models/isolation_forest.pkl")

print("\n======================")
print("üìå MODEL SUMMARY")
print("======================")

# Basic params
print("\n‚û°Ô∏è Model Parameters:")
pprint(model.get_params())

# Number of trees
print("\n‚û°Ô∏è Estimators (Trees):")
print(len(model.estimators_))

# Features used during training
print("\n‚û°Ô∏è Features Used:")
try:
    print(model.feature_names_in_)
except AttributeError:
    print("No feature_names_in_ (older sklearn version)")

# Tree depth stats
depths = [est.tree_.max_depth for est in model.estimators_]
print("\n‚û°Ô∏è Tree Depth Stats:")
print(f"Min depth: {min(depths)}")
print(f"Max depth: {max(depths)}")
print(f"Avg depth: {sum(depths)/len(depths):.2f}")

# IsolationForest offset (decision threshold)
print("\n‚û°Ô∏è Anomaly Threshold (Offset):")
print(model.offset_)

print("\n======================")
print("Done.")
print("======================")
```
